```python
def smart_near_constant_removal_optimized(
    df: pl.DataFrame,
    label_col: str = "label",
    majority_threshold: float = 0.99,
    malicious_skew_threshold: float = 0.6,
    min_minority_samples: int = 10,  # <--- CRITICAL NEW PARAMETER
    verbose: bool = True
):
    """
    Intelligently removes near-constant features, keeping only those where
    minority values are statistically significant AND strongly indicate maliciousness.
    """

    # Include Booleans! They are critical in malware (e.g., has_signature)
    feature_cols = [c for c in df.columns if c != label_col
                    and df.schema[c] in (pl.Int64, pl.Float64, pl.Boolean, pl.Int32, pl.Int16, pl.UInt8)]

    features_to_remove = []
    features_to_keep = []
    analysis_report = {}

    total_samples = df.height
    total_malicious = df.filter(pl.col(label_col) == 1).height
    baseline_malicious_rate = total_malicious / total_samples

    if verbose:
        print("=" * 70)
        print(f"üìä Analyzing {len(feature_cols)} features...")
        print(f"   Baseline Malicious Rate: {baseline_malicious_rate:.1%}")
        print(f"   Min Samples Required: {min_minority_samples}")
        print("-" * 70)

    for col in feature_cols:
        # Optimization: value_counts(sort=True) is faster than group_by().len().sort()
        vc = df[col].value_counts(sort=True)

        # Safety: Empty column check
        if vc.height == 0:
            features_to_remove.append(col)
            continue

        most_common_val = vc[0, col]
        most_common_count = vc[0, "count"]
        majority_ratio = most_common_count / total_samples

        # 1. Skip check: Is the feature diverse enough?
        if majority_ratio < majority_threshold:
            continue

        # 2. It IS near-constant. Analyze the minority.
        minority_count = total_samples - most_common_count

        # CRITICAL SAFETY CHECK: Ignore features with too few samples
        if minority_count < min_minority_samples:
            features_to_remove.append(col)
            if verbose:
                 print(f"‚ùå {col}: REMOVE (Too rare: {minority_count} samples < {min_minority_samples})")
            continue

        # 3. Analyze Label Distribution
        # Filter for rows where col is NOT the majority value
        minority_rows = df.filter(pl.col(col) != most_common_val)
        minority_malicious = minority_rows.filter(pl.col(label_col) == 1).height
        minority_malicious_rate = minority_malicious / minority_count

        # Calculate Lift (How much more dangerous is this specific value?)
        lift = minority_malicious_rate / baseline_malicious_rate if baseline_malicious_rate > 0 else 0

        # 4. Decision Logic
        if minority_malicious_rate >= malicious_skew_threshold:
            features_to_keep.append(col)
            if verbose:
                print(f"‚úÖ {col}: KEEP")
                print(f"   ‚îî‚îÄ Minority Rate: {minority_malicious_rate:.1%} (Lift: {lift:.2f}x)")
        else:
            features_to_remove.append(col)
            if verbose:
                print(f"‚ùå {col}: REMOVE (Uninformative skew: {minority_malicious_rate:.1%})")

    # Cleanup
    df_filtered = df.drop(features_to_remove)

    print("=" * 70)
    print(f"Summary: Kept {len(features_to_keep)} rare signals. Dropped {len(features_to_remove)} noise features.")

    return df_filtered, features_to_remove

# Usage
df_pdf, dropped_cols = smart_near_constant_removal_optimized(df_pdf, min_minority_samples=5)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------------------------------------
# 1. LOAD YOUR DATA
# -------------------------------------------------------
df = pd.read_csv("PDF_All_features.csv")

# -------------------------------------------------------
# 2. PLOTTING FUNCTION
# -------------------------------------------------------
def plot_binary_feature(df, feature_name, target_name='label'):
    """
    Plots a grouped bar chart for a binary feature against a target label.
    """
    plt.figure(figsize=(8, 5))

    # Create the count plot
    # x = the binary feature (0 or 1)
    # hue = the class label (Malicious or Benign)
    ax = sns.countplot(x=feature_name, hue=target_name, data=df, palette='viridis')

    plt.title(f'Distribution of "{feature_name}" by {target_name.capitalize()}', fontsize=14)
    plt.xlabel(f'{feature_name} (0 = No, 1 = Yes)', fontsize=12)
    plt.ylabel('Count of Files', fontsize=12)
    plt.legend(title=target_name.capitalize())

    # Add counts on top of bars for easier reading
    for container in ax.containers:
        ax.bar_label(container)

    plt.tight_layout()
    plt.show()

# -------------------------------------------------------
# 3. GENERATE PLOTS
# -------------------------------------------------------

# Plot 1: Encrypted
# Checks if encryption is common in malicious vs benign files
plot_binary_feature(df, 'encrypted')

# Plot 2: Uses Non-Standard Port
# Checks if weird ports are specific to malware
plot_binary_feature(df, 'openaction_count')
```

```python
import polars as pl
import numpy as np
from sklearn.feature_selection import VarianceThreshold


# ---------------------------------------------------------
# STEP 0 ‚Äî Identify numeric columns
# ---------------------------------------------------------
def get_numeric_cols(df):
    return [
        c for c, t in df.schema.items()
        if t in (pl.Int64, pl.Int32, pl.Float64, pl.Float32)
    ]


# ---------------------------------------------------------
# STEP 1 ‚Äî Variance Filtering
# Removes constant & quasi-constant features
# ---------------------------------------------------------
def variance_filter(df, threshold=0.01):
    num_cols = get_numeric_cols(df)
    X_np = df.select(num_cols).to_numpy()

    selector = VarianceThreshold(threshold)
    X_sel = selector.fit_transform(X_np)

    kept = [c for c, ok in zip(num_cols, selector.get_support()) if ok]

    print(f"Variance removed: {set(num_cols) - set(kept)}")
    return df.select(kept + [c for c in df.columns if c not in num_cols])


# ---------------------------------------------------------
# STEP 2 ‚Äî Feature‚ÄìFeature Correlation Filtering (Pearson)
# Remove redundant features with corr > threshold
# ---------------------------------------------------------
def remove_correlated_features(df, threshold=0.9):
    num_cols = get_numeric_cols(df)
    df_num = df.select(num_cols)

    corr = df_num.corr()
    to_remove = set()

    for i, c1 in enumerate(num_cols):
        for j in range(i+1, len(num_cols)):
            c2 = num_cols[j]
            # Fix: Access the correlation value directly using get_column and item(i)
            corr_val = corr.get_column(c2).item(i)

            if abs(corr_val) > threshold:
                to_remove.add(c2)

    print(f"Highly correlated removed: {to_remove}")
    return df.drop(list(to_remove))


# ---------------------------------------------------------
# STEP 3 ‚Äî Feature‚ÄìTarget Correlation Filtering
# Remove features weakly correlated with target (< min_corr)
# ---------------------------------------------------------
def target_correlation_filter(df, target, min_corr=0.05):
    num_cols = get_numeric_cols(df)
    num_cols = [c for c in num_cols if c != target]

    correlations = {
        col: df.select(pl.corr(col, target)).item()
        for col in num_cols
    }

    to_remove = [c for c, val in correlations.items() if abs(val) < min_corr]

    print(f"Weak target correlation removed: {to_remove}")
    return df.drop(to_remove)


# ---------------------------------------------------------
# STEP 4 ‚Äî Spearman Filtering (Optional)
# Rank ‚Üí Pearson = Spearman
# ---------------------------------------------------------
def spearman_filter(df, threshold=0.9):
    num_cols = get_numeric_cols(df)

    ranked = df.select([
        pl.col(c).rank().alias(f"{c}_rank")
        for c in num_cols
    ])

    rank_cols = [c + "_rank" for c in num_cols]
    corr_s = ranked.select(rank_cols).corr()

    to_remove = set()

    for i, c1 in enumerate(rank_cols):
        for j in range(i+1, len(rank_cols)):
            c2 = rank_cols[j]
            # Fix: Access the correlation value directly using get_column and item(i)
            corr_val = corr_s.get_column(c2).item(i)

            if abs(corr_val) > threshold:
                to_remove.add(c2.replace("_rank", ""))

    print(f"Spearman redundant removed: {to_remove}")
    return df.drop(list(to_remove))


# ---------------------------------------------------------
# STEP 5 ‚Äî FULL PIPELINE
# ---------------------------------------------------------
def full_feature_selection_pipeline(df, target,
                                    var_thresh=0.01,
                                    pearson_thresh=0.9,
                                    target_corr_thresh=0.05,
                                    use_spearman=False,
                                    spearman_thresh=0.9):

    print("\nüîπ STEP 1 ‚Äî VARIANCE FILTERING")
    df1 = variance_filter(df, var_thresh)

    print("\nüîπ STEP 2 ‚Äî PEARSON FEATURE-FEATURE CORRELATION")
    df2 = remove_correlated_features(df1, pearson_thresh)

    print("\nüîπ STEP 3 ‚Äî FEATURE‚ÄìTARGET CORRELATION")
    df3 = target_correlation_filter(df2, target, target_corr_thresh)

    if use_spearman:
        print("\nüîπ STEP 4 ‚Äî SPEARMAN FILTER (Optional)")
        df4 = spearman_filter(df3, spearman_thresh)
        return df4

    return df3
```
